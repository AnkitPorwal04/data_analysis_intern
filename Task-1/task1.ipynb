## Identifying Outliers and addressing them

```python
df3 = pd.read_csv('D:\A\Programming\Ipynb\duplicateANDnull_removed.csv')
```

```python
# making two dataset one for numerical and one for non numerical values so we can apply outlier addressing
numeric_columns = df3.select_dtypes(include=['number'])
categorical_columns = df3.select_dtypes(exclude=['number'])
numeric_columns
```

```python
Q1 = numeric_columns.quantile(0.25)
Q3 = numeric_columns.quantile(0.75)
IQR = Q3 - Q1
threshold = 1.5
outliers = ((numeric_columns < (Q1 - threshold * IQR)) | (numeric_columns > (Q3 + threshold * IQR))).any(axis=1)
print(IQR)
```

```python
numeric_columns = numeric_columns[~outliers]
```

```python
df_3 = pd.concat([categorical_columns, numeric_columns], axis=1)
```

```python
df_3
```

```python
df3.to_csv('outliersremoved.csv', index=False)
```

## Scaling and Encoding


```python
df4 = pd.read_csv('D:\A\Programming\Ipynb\outliersremoved.csv')
```

```python
numeric_columns = df4.select_dtypes(include=['number'])
categorical_columns = df4.select_dtypes(exclude=['number'])
```

```python
from sklearn.preprocessing import MinMaxScaler

# Instantiate the MinMaxScaler
scaler = MinMaxScaler()

# Scale all numeric columns
numeric_columns_scaled = scaler.fit_transform(numeric_columns)

# Create a DataFrame from the scaled numeric columns
df4_scaled = pd.DataFrame(numeric_columns_scaled, columns=numeric_columns.columns)
```

```python
df4_scaled
```

```python
categorical_columns = pd.get_dummies(categorical_columns, drop_first=True)
```

```python
# Concatenate the scaled numeric columns with the original categorical columns
df4_scaled = pd.concat([categorical_columns, df4_scaled], axis=1)
```

```python
df4_scaled.to_csv('final_dataset.csv', index=False)
```
